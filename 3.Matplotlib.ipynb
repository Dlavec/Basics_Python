# LIBRERIAS CIENTIFICAS
Las librerías son un conjunto de funciones y sirven como agregado a las funciones que ya tenemos inicialmente. Algunas librerías útiles son:

### *Cálculo numérico:
Numpy,
Scipy

### *Gráficos:
Matplotlib

### *Otros:
Sympy (lenguaje simbólico),
Pandas (Análisis de datos)

## MATPLOTLIB

```python
import numpy as np          #Importo las libreria y les asigno un nombre
import matplotlib.pyplot as plt
x = np.linspace(0,4*np.pi,100)

y = np.sin(x)

z = np.cos(x)
plt.plot(x,y) #Esta es la forma mas simple de graficar
plt.plot(x,z) #Grafica ambos plots juntos
```

```python
x = np.linspace(0,4*np.pi,100)

y = np.sin(x)

z = np.cos(x)

plt.figure()
plt.plot(x, y, '-o', color='blue', label='sin(x)')  #Elijo colores y ejes
plt.plot(x, z, '-s', color='red', label='cos(x)')
plt.grid()                                          #Agrego grilla
plt.xlabel('x')                                     #TItulos de los ejes
plt.ylabel('y')
plt.title('Funciones Trigonométricas')              #Titulo principal
plt.legend()                                        #Leyendas sin(x) y cos(x)
plt.show()
```

```python
##Otra forma de graficar
Creo la figura y el gráfico como si fueran variables (se les llama objetos) y después le cambio los atributos
```

```python
fig,ax = plt.subplots(1) #Creo estas "variables"
#Le cambio atributos
ax.grid(True)
ax.plot(x, y, color = 'b', label = 'sen(x)')
```
